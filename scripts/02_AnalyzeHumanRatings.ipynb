### Compute ICC

```python
df1 = pd.read_csv('../data/05_human_ratings/human_ratings_part1.csv')
df1 = df1.drop(columns=['age', 'sex'])
df1.head()
```

```python
ratings_matrix1 = df1.T.iloc[:, :]
icc_value1 = compute_icc(ratings_matrix1)
icc_value1
```

```python
df2 = pd.read_csv('../data/05_human_ratings/human_ratings_part2.csv')
df2 = df2.drop(columns=['age', 'sex'])
df2.head()
```

```python
ratings_matrix2 = df2.T.iloc[:, :]
icc_value2 = compute_icc(ratings_matrix2)
icc_value2
```

### Combine Means and Save

```python
df1_mean = df1.mean()
df2_mean = df2.mean()

df1_mean_df = df1_mean.reset_index()
df1_mean_df.columns = ["Speech", "Mean_Score"]

df2_mean_df = df2_mean.reset_index()
df2_mean_df.columns = ["Speech", "Mean_Score"]

df_human_ratings = pd.concat([df1_mean_df, df2_mean_df], ignore_index=True)

df_human_ratings.to_csv('../data/05_human_ratings/averaged_human_ratings.csv')
df_human_ratings.head()
```
